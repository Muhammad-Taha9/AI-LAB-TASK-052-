```python
def fizz_buzz():
    print("Welcome to Fizz Buzz!")
    print("Let's play the first 100 rounds:")
    
    for number in range(1, 101):  
        if number % 3 == 0 and number % 5 == 0:
            print("Fizz Buzz")
        elif number % 3 == 0:  
            print("Fizz")
        elif number % 5 == 0:  
            print("Buzz")
        else:  
            print(number)

fizz_buzz()
```
